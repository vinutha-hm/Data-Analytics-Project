```python
df['Yield'].isnull().sum()
```

```python
df = df.replace([float('inf'), -float('inf')], pd.NA)
df = df.dropna(subset=['Yield'])
```

```python
print(df['Yield'].describe())
```

```python
df.head()
```

```python
# Group by Year and Crop, then count unique states
multi_state_crops = (
    df.groupby(['Crop_Year', 'Crop'])['State_Name']
    .nunique()
    .reset_index()
    .rename(columns={'State_Name': 'State_Count'})
)

# Filter where crop is grown in more than 1 state in the same year
multi_state_crops = multi_state_crops[multi_state_crops['State_Count'] > 1]

print(multi_state_crops.head())
```

```python
filtered_df = df.merge(multi_state_crops[['Crop_Year', 'Crop']], on=['Crop_Year', 'Crop'], how='inner')

print(filtered_df[['State_Name', 'Crop_Year', 'Crop']].drop_duplicates().head(20))
```
